```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(3):
    plt.figure()
    F1,X1,Y1 = np.loadtxt(f'Peak{i+1}',skiprows = 1).T
    plt.plot(F1,X1)
    plt.title(f'X: Peak {i+1}')
    plt.show()
    plt.plot(F1,Y1)
    plt.title(f'Y: Peak {i+1}')
    plt.show()
```

```python
F1,X1,Y1 = np.loadtxt('Peak1',skiprows = 1).T
F1_mean = np.mean(F1.reshape(100,20),axis=-1)
X1_mean = np.mean(X1.reshape(100,20),axis=-1)
X1_std = np.std(X1.reshape(100,20),axis=-1)

Y1_mean = np.mean(Y1.reshape(100,20),axis=-1)
Y1_std = np.std(Y1.reshape(100,20),axis=-1)

plt.errorbar(F1_mean,X1_mean,yerr = X1_std)
plt.show()
plt.errorbar(F1_mean,Y1_mean,yerr = Y1_std)
```
